### One-Hot Encoding for Categorical Variables

One-Hot Encoding is a technique used to convert categorical variables into a numerical format that can be provided to machine learning algorithms. Each category is transformed into a new binary column (0 or 1).

We will apply this specifically to categorical columns that have less than 5 unique categories to avoid creating too many new features, which can lead to high dimensionality.

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
cols_to_encode = []
for col in categorical_cols:
    num_unique = df[col].nunique()
    if num_unique < 5:
        cols_to_encode.append(col)
```

```python
if cols_to_encode:
    df_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=False)
    display(df_encoded.head())
    df.shape
    df_encoded.shape
    df = df_encoded
else:
    print("no less than 5 unique")
```

```python
from sklearn.preprocessing import LabelEncoder
remaining_categorical_cols = df.select_dtypes(include='object').columns.tolist()
le = LabelEncoder()
for col in remaining_categorical_cols:
    df[col] = le.fit_transform(df[col])
display(df.head())
df.info()
```
